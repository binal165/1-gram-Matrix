```python
import nltk 
import pickle
from collections import Counter
import pandas as pd
```

```python
stopwords = ["all","just","being","over","both","through","yourselves","its","before","herself","had","should","to","only",
             "under","ours","has","do","them","his","very","they","not","during","now","him","nor","did","this","she","each",
             "further","where","few","because","doing","some","are","our","ourselves","out","what","for","while","does","above",
             "between","t","be","we","who","were","here","hers","by","on","about","of","against","s","or","own","into",
             "yourself", "down","your","from","her","their","there","been","whom","too","themselves","was","until","more",
             "himself","that","but","don","with","than","those","he","me","myself","these","up","will","below","can","theirs",
             "my","and","then","is","am","it","an","as","itself","at","have","in","any","if","again","no","when","same","how",
             "other","which","you","after","most","such","why","a","off","i","yours","so","the","having","once"]

file = open("abstracts.txt")
text = file.readlines()
```

```python
# Remove stop words
data =[]
for line in text:
    splt = line.split()
    if splt[1] != 'null':
        free = [w for w in splt if not w in stopwords]
        data.append(free)
```

```python
# Creating a set of unique words in all the documents
allWords = []
allDocs = []

allDocsWordCount = {}
fullList = []

for doc in data:
    # Making a list of all the document ids
    allDocs.append(doc[0])
    #docWordCount.append(doc[0])
    docWordCount = {}
    for word in doc[1:]:
        # Creating a list of all the words to eventually make a unique list
        allWords.append(word)
        # Get the frequencies of each word
        if word not in docWordCount:
            docWordCount[word] = 1
        else:
            docWordCount[word] += 1
    fullList.append({doc[0]:docWordCount})

# All unique words
uniqueWords = list(set(allWords))
# Remove numeric values
uniqueWords = [x for x in uniqueWords if not any(c.isdigit() for c in x)]

# For every word, iterate the list of document dictionaries
# Check if the word exists for each document
# If the word exists, use the frequency else append 0
rowDict={}
for word in uniqueWords:
    for doc in fullList:
        x = doc.values()
        if word not in rowDict:
            rowDict[word]=[]
        if word in x[0].keys():
            rowDict[word].append(x[0][word])
        else:
            rowDict[word].append(0)
            
# Creating a dataframe
df = pd.DataFrame(rowDict, index=allDocs)
df
```
